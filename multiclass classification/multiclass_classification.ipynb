```python
testing_data=np.array(dataset1)
y_test=np.array(y_test)
```

```python
testing_data.shape
```

```python
y_test=np.array(y_test)
y_test.shape
```

```python
y_test = to_categorical(y_test, num_classes=4)
```

```python
y_test
```

```python
y_test.shape
```

```python
resnet=ResNet50(weights='imagenet',include_top='False')
#eNet=EfficientNetB4(weights='imagenet',include_top='False')
```

```python
for layer in resnet.layers:
  layer.trainable = False
```

```python
predictions = Dense(4, activation='softmax')(resnet.layers[-2].output)
model = Model(inputs=resnet.input, outputs=predictions)
```

```python
opt = keras.optimizers.Adam(learning_rate=0.0001)
model.compile(loss='categorical_crossentropy',optimizer='Adam', metrics=['categorical_accuracy'])
```

```python
model.summary()
```

```python
from keras.callbacks import ModelCheckpoint,EarlyStopping
filepath="/content/drive/MyDrive/Saved_models/Resnet_multiclass_onehotencoded/weights-improvement-{epoch:02d}-{categorical_accuracy:.2f}.hdf5"
checkpoints=ModelCheckpoint(filepath,monitor='val_categorical_accuracy',verbose=1,save_best_only=True,mode='max')
earlystop=EarlyStopping(monitor='val_loss',patience=30,verbose=1)
callbacks_list=[checkpoints,earlystop]
```

```python
history=model.fit(training_data,y_train,validation_data=(testing_data,y_test),epochs=400,callbacks=callbacks_list,batch_size=50,shuffle=True)
```

```python
train_loss,train_acc=model.evaluate(training_data,y_train)
test_loss,test_acc=model.evaluate(testing_data,y_test)
print('Train: %.3f,Test: %.3f' %(train_acc,test_acc))
```

```python
#loss
plt.plot(history.history['loss'], label='train loss')
plt.plot(history.history['val_loss'], label='val loss')
plt.legend()
plt.show()

#accuracies
plt.plot(history.history['categorical_accuracy'], label='train acc')
plt.plot(history.history['val_categorical_accuracy'], label='val acc')
plt.legend()
plt.show()
```
